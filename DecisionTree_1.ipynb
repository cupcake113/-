<h1>نموذج يصنف زهرة ايرس ويتنباء بانوعها بناء على صفات محددة</h1>

```python
# الخلية 1: استيراد المكتبات وتحميل البيانات
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# تحميل بيانات Iris
iris = load_iris()
X = iris.data  # الميزات
y = iris.target  # الأهداف (الفئات)
print(iris)
```

```python
# الخلية 2: إنشاء وتدريب نموذج شجرة القرار
from sklearn.tree import DecisionTreeClassifier

# إنشاء نموذج شجرة قرار
model = DecisionTreeClassifier()
model.fit(X, y)  # تدريب النموذج
```

```python
# الخلية 3: عرض الفئات الخاصة بالنموذج
print("Classes in the model:", model.classes_)
```

```python
# الخلية 4: إنشاء الرسم البياني لشجرة القرار
from sklearn.tree import plot_tree

plt.figure(figsize=(12, 6))
plot_tree(model,
          max_depth=2,  # عرض الشجرة حتى عمق 2 فقط
          fontsize=10,
          feature_names=iris.feature_names,  # أسماء الميزات
          class_names=iris.target_names.tolist()  # تحويل الأسماء إلى قائمة
         )
plt.show()  # عرض الرسم البياني
```
